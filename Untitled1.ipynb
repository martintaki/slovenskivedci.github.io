```python

import oyaml as yaml
import glob
import copy

countries={}
institutions={}
alllst=[]

from urllib.parse import unquote

order = ["name","last","year",
"sex","position","field",
"affiliation",
"city",
"country","hindex","scholar","last_update","links","img"]

 
for y in glob.glob("./people/*.yaml"):
    print(y) 
    with open(y) as f:
        dic = yaml.safe_load(f)
    
    dicnew = copy.deepcopy(dic)


    for key in order:
        val = dicnew[key]
        del dicnew[key]
        dicnew[key] = val

    with open(y, 'w') as file:
        documents = yaml.dump(dicnew, file)
    

```
