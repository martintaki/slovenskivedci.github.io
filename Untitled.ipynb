```python
import numpy as np

import seaborn as sns
import oyaml as yaml
import os
import matplotlib.pyplot as plt
import pandas as pd
import unicodedata

def repl(text):
	return unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode('UTF-8').replace(",","_")
```

```python
field = [ e["field"] for e in alldata]
hindex = [ e["hindex"] for e in alldata]
sex = [ e["sex"] for e in alldata]
affiliation = [ e["affiliation"] for e in alldata]
country = [ e["country"] for e in alldata]

df = pd.DataFrame(data={"field":field,"hindex":hindex,
                        "affiliation":affiliation,
                        "country":country,
                        "sex":sex})
```

```python

sns.set_theme(style="whitegrid", palette="muted")

# Draw a categorical scatterplot to show each observation
ax = sns.swarmplot(data=df, x="hindex", y="sex")# "species"
ax.set(ylabel="")
```

```python

fig = plt.figure(figsize=(20, 5), dpi=100)
ax = fig.add_axes([0,0,1,1])

d = dic["affiliation"]
d = sorted([(k,d[k]) for k in d],  key = lambda e: (-e[1],repl(e[0])))


langs = [e[0] for e in d]
students = [e[1] for e in d]
ax.bar(langs,students)
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees
plt.grid(True)
plt.tight_layout()


plt.savefig("stats/affiliation.jpg", bbox_inches = 'tight')
```
