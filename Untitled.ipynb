```python
import numpy as np


import yaml
import os
import matplotlib.pyplot as plt

import unicodedata

def repl(text):
	return unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode('UTF-8').replace(",","_")
```

```python
with open("_data/page.yaml") as f:
		dic = yaml.safe_load(f)
```

```python



plt.hist(dic["hindex"],40)
plt.grid(True)

plt.tight_layout()
plt.savefig("stats/hindex.jpg")
```

```python

fig = plt.figure(figsize=(20, 5), dpi=100)
ax = fig.add_axes([0,0,1,1])

d = dic["affiliation"]
d = sorted([(k,d[k]) for k in d],  key = lambda e: (-e[1],repl(e[0])))


langs = [e[0] for e in d]
students = [e[1] for e in d]
ax.bar(langs,students)
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees
plt.grid(True)
plt.tight_layout()

plt.savefig("stats/affiliation.jpg", bbox_inches = 'tight')
```

```python
fig = plt.figure(figsize=(20, 5), dpi=100)
ax = fig.add_axes([0,0,1,1])

d = dic["countries"]
d = sorted([(k,d[k]) for k in d],  key = lambda e: (-e[1],repl(e[0])))


langs = [e[0] for e in d]
students = [e[1] for e in d]
ax.bar(langs,students)
plt.xticks(rotation = 90) # Rotates X-Axis Ticks by 45-degrees
plt.grid(True)
plt.tight_layout()

plt.savefig("stats/countries.jpg", bbox_inches = 'tight')
```
